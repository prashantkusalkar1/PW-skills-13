Question 1 : Explain the fundamental differences between DDL, DML, and DQL commands in SQL. Provide one example for each type of command.

1. DDL (Data Definition Language)

DDL commands are used to define or modify the structure of the database (tables, schemas, etc.).

They deal with schema changes, not actual data.
Changes are usually permanent (auto-committed).

Examples of DDL commands:

CREATE
ALTER
DROP
TRUNCATE

```python
CREATE TABLE Employees (
    ID INT,
    Name VARCHAR(50),
    Salary DECIMAL(10,2)
);
```

2. DML (Data Manipulation Language)

DML commands are used to manipulate the data inside tables.

They affect rows (records) in the table.
Changes can be rolled back (if transactions are used).

Examples of DML commands:

INSERT
UPDATE
DELETE

```python
INSERT INTO Employees (ID, Name, Salary)
VALUES (1, 'John', 50000);
```

3. DQL (Data Query Language)

DQL is used to retrieve data from the database.

It focuses only on querying (reading) data.

Main command:

SELECT

```python
SELECT * FROM Employees;
```

Question 3 : Explain the difference between LIMIT and OFFSET clauses in SQL. How
would you use them together to retrieve the third page of results, assuming each page
has 10 records?

LIMIT
Specifies how many rows to return
Think of it as: “Give me this many records”

Example:

SELECT * FROM Employees
LIMIT 10;

OFFSET
Specifies how many rows to skip before starting to return results
Think of it as: “Skip these many records first”

Example:

SELECT * FROM Employees
OFFSET 10;

Using LIMIT + OFFSET together (Pagination)

Formula:

OFFSET = (page_number - 1) × records_per_page

Example: Retrieve 3rd page, 10 records per page
Page number = 3
Records per page = 10
OFFSET = (3 - 1) × 10 = 20

Query:

SELECT * FROM Employees
LIMIT 10 OFFSET 20;

Question 4 : What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage

A Common Table Expression (CTE) is a temporary result set in SQL that you can define using the WITH clause and then reference within a SELECT, INSERT, UPDATE, or DELETE query.

Think of it like a temporary named query that makes complex SQL easier to read and manage.

```python
WITH HighSalaryEmployees AS (
    SELECT Name, Salary
    FROM Employees
    WHERE Salary > 50000
)
SELECT *
FROM HighSalaryEmployees;
```

Question 5 : Describe the concept of SQL Normalization and its primary goals. Briefly
explain the first three normal forms (1NF, 2NF, 3NF).

Normalization is the process of organizing data in a database to:

Reduce data redundancy (duplicate data)
Improve data integrity
Make the database more efficient and consistent

1NF (First Normal Form)
Each column contains atomic (single) values
No repeating groups or multiple values in one column

❌ Not in 1NF:

ID	Phone Numbers
1	9876, 1234

✅ In 1NF:

ID	Phone Number
1	9876
1	1234

2NF (Second Normal Form)
Must be in 1NF
All non-key columns must depend on the entire primary key (no partial dependency)

Scenario:
If a table has a composite key, no column should depend on only part of it.

3NF (Third Normal Form)
Must be in 2NF
No transitive dependency (non-key columns should not depend on other non-key columns)

Question 7 : Generate a report showing CustomerName, Email, and the
TotalNumberofOrders for each customer. Include customers who have not placed
any orders, in which case their TotalNumberofOrders should be 0. Order the results
by CustomerName.

✅ SQL Query
SELECT
    c.CustomerName,
    c.Email,
    COUNT(o.OrderID) AS TotalNumberOfOrders
FROM Customers c
LEFT JOIN Orders o
    ON c.CustomerID = o.CustomerID
GROUP BY
    c.CustomerID, c.CustomerName, c.Email
ORDER BY
    c.CustomerName;


    Explanation
LEFT JOIN → ensures all customers are included (even without orders)
COUNT(o.OrderID) → counts only matching orders (returns 0 if none)
GROUP BY → groups results per customer
ORDER BY → sorts alphabetically by CustomerName

Question 8 : Retrieve Product Information with Category: Write a SQL query to
display the ProductName, Price, StockQuantity, and CategoryName for all
products. Order the results by CategoryName and then ProductName alphabetically.

SQL Query
SELECT
    p.ProductName,
    p.Price,
    p.StockQuantity,
    c.CategoryName
FROM Products p
INNER JOIN Categories c
    ON p.CategoryID = c.CategoryID
ORDER BY
    c.CategoryName ASC,
    p.ProductName ASC;

    Explanation
INNER JOIN → connects products with their respective categories
p.ProductName, p.Price, p.StockQuantity → product details
c.CategoryName → category name from Categories table
ORDER BY:
First sorts by CategoryName
Then sorts by ProductName alphabetically within each category

Question 9 : Write a SQL query that uses a Common Table Expression (CTE) and a
Window Function (specifically ROW_NUMBER() or RANK()) to display the
CategoryName, ProductName, and Price for the top 2 most expensive products in
each CategoryName.

```python
WITH RankedProducts AS (
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        ROW_NUMBER() OVER (
            PARTITION BY c.CategoryName
            ORDER BY p.Price DESC
        ) AS rank_num
    FROM Products p
    INNER JOIN Categories c
        ON p.CategoryID = c.CategoryID
)
SELECT
    CategoryName,
    ProductName,
    Price
FROM RankedProducts
WHERE rank_num <= 2
ORDER BY
    CategoryName,
    Price DESC;
```

Alternative (using RANK)**

Use RANK() instead if you want to handle ties (same price):

RANK() OVER (PARTITION BY c.CategoryName ORDER BY p.Price DESC)

👉 Difference:

ROW_NUMBER() → always gives exactly 2 rows per category
RANK() → may return more if prices are tied

1. Top 5 Customers by Total Spending

```python
SELECT
    c.first_name,
    c.last_name,
    c.email,
    SUM(p.amount) AS total_spent
FROM customer c
JOIN payment p
    ON c.customer_id = p.customer_id
GROUP BY
    c.customer_id, c.first_name, c.last_name, c.email
ORDER BY
    total_spent DESC
LIMIT 5;
```
